### Sorting data

```python
import pandas as pd
```

Sorting data in pandas involves arranging the data in a DataFrame or Series based on the values of one or more columns or indices. This can be done either in ascending or descending order.

```python
people = {
    'first': ['John', 'Jane', 'Jim','Adam'],
    'last': ['Doe', 'Doe', 'Brown', 'Doe'],
    'email': ['john.doe@example.com', 'jane.doe@example.com', 'jim.brown@example.com','adam.doe@example.com']
}
```

```python
people_df = pd.DataFrame(people) 
people_df
```

```python
# Sort by last name
people_df.sort_values(by = 'last', ascending = True)
```

```python
# Sort by last name and first name
people_df.sort_values(by = ['last','first'], ascending = [True, True], inplace = True)
people_df
```

```python
# Resetting indexes
#people_df = people_df.reset_index(drop=True)
```

```python
# Sort by index
people_df.sort_index()
```

```python
# Series objects also have a sort method
people_df['first'].sort_values()
```
