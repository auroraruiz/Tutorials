```python
import requests
from bs4 import BeautifulSoup
from io import StringIO

# URL of the webpage
url = 'https://www.itsrugby.co.uk/rugby-top14.html'

# Fetch the HTML content
response = requests.get(url)
soup = BeautifulSoup(response.text, 'lxml')

# Find the specific table using BeautifulSoup
table = soup.select_one('div:nth-of-type(10) > div:nth-of-type(2) > div > div:nth-of-type(1) > table:nth-of-type(3)')

# Convert the table to a string and wrap it in a StringIO object
table_html = str(table)
table_data = StringIO(table_html)

# Parse the table with pandas
df = pd.read_html(table_data, header = 0, index_col = 'Teams')[0]

# Get the index of the last row
last_index = df.index[-1]

# Drop the last row
df = df.drop(index=last_index)

# Cleaning empty columns
df.dropna(axis = 'columns', how = 'all', inplace = True)

df.head(20)
```
