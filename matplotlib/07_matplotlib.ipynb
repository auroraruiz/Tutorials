```python
from matplotlib import pyplot as plt
```

```python
import pandas as pd
```

```python
# To reset to default configuration
plt.rcdefaults()
```

```python
# To see all available styles run:
# plt.style.available
plt.style.use('seaborn-v0_8')
```

```python
x = [5, 7, 8, 5, 6, 7, 9, 2, 3, 4, 4, 4, 2, 6, 3, 6, 8, 6, 4, 1]
y = [7, 4, 3, 9, 1, 3, 2, 5, 2, 4, 8, 7, 1, 6, 4, 9, 7, 7, 5, 1]
```

```python
plt.scatter(x,y, s=50, c = 'green', marker = 'x')
```

```python
# Adding edges to our sacatter plot
plt.scatter(x,y, s=50, c = 'green', edgecolors='black', linewidth = 1,alpha= 0.5)
```

```python
# Colors and sizes can be assigned per point
```

```python
# Numbers between 1 and 10 (could correspond to answers on a survey)
colors = [7, 5, 9, 7, 5, 7, 2, 5, 3, 7, 1, 2, 8, 1, 9, 2, 5, 6, 7, 5]
```

```python
plt.scatter(x,y, s=50, c = colors, cmap='Blues', 
            edgecolors='black', linewidth = 1,alpha= 0.5)
```

```python
import matplotlib.cm as cm

scatter = plt.scatter(x,y, s=50, c = colors, cmap=cm.hot.reversed(),
                      edgecolors='black', linewidth = 1,alpha= 0.5)

cbar = plt.colorbar(scatter)
cbar.set_label('Satisfaction')

plt.show()
```

If you ever want to make a discrete color bar, go [here](https://stackoverflow.com/questions/8342549/add-colorbar-to-a-sequence-of-line-plots)

```python
# We can also change the sizes of the points(to represent population for example)
sizes = [209, 486, 381, 255, 191, 315, 185, 228, 174,
        538, 239, 394, 399, 153, 273, 293, 436, 501, 397, 539]
```

```python
import matplotlib.cm as cm

scatter = plt.scatter(x,y, s=sizes, c = colors, cmap=cm.hot.reversed(),
                      edgecolors='black', linewidth = 1,alpha= 0.5)

cbar = plt.colorbar(scatter)
cbar.set_label('Satisfaction')

plt.show()
```

Let's now plot some real world data from an .csv file.

The .csv has data about the top 200 trending videos on Youtube on a given day.

We want to create a scatter plot that shows the number of views and the number of likes these videos received.

```python
data = pd.read_csv('data/2019-05-31-data.csv')
data.head()
```

```python
data.shape
```

```python
view_count = data['view_count']
likes = data['likes']
ratio = data['ratio'] # Ratio of likes vs dislikes

plt.scatter(view_count,likes)
```

```python
view_count = data['view_count']
likes = data['likes']
ratio = data['ratio'] # Ratio of likes vs dislikes

plt.scatter(view_count,likes)
```

```python
scatter = plt.scatter(view_count,likes, c = ratio, cmap='summer',
                      edgecolors='black', linewidth = 1,alpha= 0.75)

plt.xscale('log') # Helps you visualize the data
plt.yscale('log')

cbar = plt.colorbar(scatter)
cbar.set_label('Ratio')

plt.title('Trending Youtube Videos on 2019-05-31')
plt.xlabel('View Count')
plt.ylabel('Total Likes')

plt.show()
```
